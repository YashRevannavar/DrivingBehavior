### Accel Euclidean Distance

```python
def euclideanDistance(df):
    """
    This function converts Accel_X, Accel_Y and Accel_Z colmun to an euclidean column as per euclidean Distance formula.
    parma df: Input as DataFrame.
    type df: pandas dataframe
    return df: output as DataFrame including euclidean column.
    type df: pandas dataframe
    """
    D = []
    for i in range(len(df)-1):
        D.append((((df.Accel_X[i+1]- df.Accel_X[i])**2 )+((df.Accel_Y[i+1]- df.Accel_Y[i])**2 )+((df.Accel_Z[i+1]- df.Accel_Z[i])**2 )).round(4))
    D.append(0)
    df['euclidean'] = D
    return df
dfC = euclideanDistance(C16)
dfN = euclideanDistance(N28)
dfA = euclideanDistance(A41)
```

### Statical Analisis

#### Data processing for statical graphs

```python
# function to calculate mean, median, max, min, Std and count of  Accel_X, Accel_Y, Accel_Z and Velocity.

def statsDf(df: pd.DataFrame) -> np.empty([24]):
    """
    This function generates a array of statisical data such as mean, median, max, min, standard deviation and count 
        for the accel x, accel y accel z and velocity.
    parma df: Input as DataFrame.
    type df: pandas dataframe
    return arrayData: array consisting of the statistical data.
    type arrayData: numpy array
    """
    
    arrayData = np.empty([24])
    arrayData[0] = df.Accel_X.mean()
    arrayData[1] = df.Accel_X.median()
    arrayData[2] = df.Accel_X.max()
    arrayData[3] = df.Accel_X.min()
    arrayData[4] = df.Accel_X.std()
    arrayData[5] = df.Accel_X.count()

    arrayData[6] = df.Accel_Y.mean()
    arrayData[7] = df.Accel_Y.median()
    arrayData[8] = df.Accel_Y.max()
    arrayData[9] = df.Accel_Y.min()
    arrayData[10] = df.Accel_Y.std()
    arrayData[11] = df.Accel_Y.count()

    arrayData[12] = df.Accel_Z.mean()
    arrayData[13] = df.Accel_Z.median()
    arrayData[14] = df.Accel_Z.max()
    arrayData[15] = df.Accel_Z.min()
    arrayData[16] = df.Accel_Z.std()
    arrayData[17] = df.Accel_Z.count()

    arrayData[18] = df.Velocity.mean()
    arrayData[19] = df.Velocity.median()
    arrayData[20] = df.Velocity.max()
    arrayData[21] = df.Velocity.min()
    arrayData[22] = df.Velocity.std()
    arrayData[23] = df.Velocity.count()
    arrayData = arrayData[np.newaxis,:]
    return arrayData
```
